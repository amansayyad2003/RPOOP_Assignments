```python
class Bank_Account:
    def __init__(self, depositor, account_numer, account_type, bank_balance, passcode):
        self.depositor = depositor
        self.account_numer = account_numer
        self.account_type = account_type
        self.bank_balance = bank_balance
        self.passcode = passcode

    
    def get_depositor(self):
         return self.depositor
    
    def set_depositor(self, depositor):
        self.depositor = depositor
        
    def get_account_number(self):
        return self.account_numer
    
    def set_account_number(self, account_number):
        self.account_number = account_number
    
    def get_account_type(self):
        return self.account_type
    
    def set_account_type(self, account_type):
        self.account_type = account_type
        
    def get_bank_balance(self):
        return self.bank_balance
    
    def get_passcode(self):
        return self.passcode
    
    def set_passcode(self, passcode):
        self.passcode = passcode
        
    def set_depositor(self, depositor):
        self.depositor = depositor
        return
    
    def withdraw(self, amount):
        if(self.bank_balance < amount):
            self.bank_balance = 0
        else:
            self.bank_balance = self.bank_balance - amount
        return
    
    def deposit(self, amount):
        self.bank_balance = self.bank_balance + amount
        return
    
    
            
            
            
```
