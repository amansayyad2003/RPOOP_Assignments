```python
class Human:
    def __init__(self, name, age):
        self.__name = name
        self.__age = age
        
    def __str__(self, name, age):
        return f"{self.name}({self.age})"
    
    def get_name(self):
        return self.__name
    
    def set_name(self, name):
        self.__name = name
        
    def get_age(self):
        return self.__age
    
    def set_age(self, age):
        self.__age = age
```

```python
class programmer(Human):
    def __init__(self, name, age, language):
        Human.__init__(self, name, age)
        self.__language = language
    
    def __str__(self, name, age, language):
        return f"{self.name}({self.age, self.language})"

    def get_language(self):
        return self.__language
    
    def set_language(self, language):
        self.__language = language
        
    
```

```python
class doctor(Human):
    def __init__(self, name, age, qualification):
        super().__init__(name, age)
        self.__qualification = qualification
        
    def __str__(self, name, age):
        return f"{self.name}({self.age, self.qualification})"

    def get_qulification(self):
        return self.__qualification
    
    def set_qualification(self, qualification):
        self.__qualification = qualification
```

```python
class Software_Engineer(programmer):
    def __init__(self, name, age, language, company):
        super().__init__(name, age, language)
        self.__company = company
        
    def __str__(self, name, age, language):
        return f"{self.name}({self.age, self.language, self.company})"
    
    def get_company(self):
        return self.__company
    
    def set_company(self, company):
        self.__company = company
```

```python
class student(Human):
    def __init__(self, name, age, studying_year, college_name):
        Human.__init__(self, name, age)
        self.__studying_year = studying_year
        self.__college_name = college_name
    
    def __str__(self, name, age, studying_year):
        return f"{self.name}({self.age, self.studying_year})"

    def get_studying_year(self):
        return self.__studying_year
    
    def set_studying_year(self, studying_year):
        self.__studying_year = studying_year
    
    def get_college_name(self):
        return self.__college_name
    
    def set_college_name(self, college_name):
        self.__college_name = college_name
        
```

```python
class hacker(programmer, student):
    def __init__(self, name, age, language, studying_year, college_name, type_of_hacker):
        #super().__init__(name, age, language, studying_year, college_name)
        programmer.__init__(self, name, age, language)
        student.__init__(self, name, age, studying_year, college_name)
        self.__type_of_hacker = type_of_hacker
    
    def get_type_of_hacker(self):
        return self.__type_of_hacker
    
    def set_type_of_hacker(self, type_of_hacker):
        self.__type_of_hacker = type_of_hacker
        
```

```python
def student_attr_fun(student):
    print("Choose the Functions you want to access")
    while(True):
        num = int(input("\t1. Get studying Year\n\t2. Get College Name\n\t3. Set Studying Year\n\t4. Set College Name\n\t5. Exit\n"))
        if(num == 1):
            print("Studying Year: ", student.get_studying_year())
        elif(num == 2):
            print("College Name: " +  student.get_college_name())
        elif(num == 3):
            study_year = int(input("Enter the new studying year: "))
            student.set_studying_year(study_year)
            print("Year updated\n")
        elif(num == 4):
            college_name = input("Enter new college name: ")
            student.set_college_name(college_name)
            print("College Name updated\n")
        else:
            break
    
```

```python
def Human_attr_fun(human):
    print("Choose the functions you want to access")
    while(True):
        num = int(input("\t1. Get Name\n\t2. Get Age\n\t3. Set Name\n\t4. Set Age\n\t5. Exit"))
        if(num == 1):
            print("Name: " + human.get_name())
        elif(num == 2):
            print("Age: ", human.get_age())
        elif(num == 3):
            name = input("Enter the new Name: ")
            human.set_name(name)
            print("Name is updated !!\n")
        elif(num == 4):
            age = input("Enter the new age")
            human.set_age(age)
            print("Age is updated !!")
        else:
            break
```
