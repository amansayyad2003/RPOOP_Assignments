```python
class Human:
    def __init__(self, name, age):
        self.__name = name
        self.__age = age
        
    def __str__(self, name, age):
        return f"{self.name}({self.age})"
    
    def get_name(self):
        return self.__name
    
    def set_name(self, name):
        self.__name = name
        
    def get_age(self):
        return self.__age
    
    def set_age(self, age):
        self.__age = age

        
```
